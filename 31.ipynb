```python
def hello(count):
  if count == 0:
    return
  print('Hello, world!', count)
  count-=1
  hello(count)

hello(5)
```

```python
def factorial(n):
  if n == 1:
    return 1
  return n*factorial(n-1)
print(factorial(5))
```

```python
def fib(n):
  if n <=  1:
    return n
  else:
    return fib(n-1)+fib(n-2)
n = int(input())
print(fib(n))
```
